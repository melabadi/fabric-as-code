# Demo load notebook
Provisioned by **fabric-as-code**. Writes a small demo dataset into the attached Lakehouse as a Delta table.

```python
from pyspark.sql import Row
from datetime import date

rows = [
    Row(order_id=1, customer_id=101, order_date=date(2026, 1, 5), amount=120.50, status='shipped'),
    Row(order_id=2, customer_id=102, order_date=date(2026, 1, 6), amount=75.00,  status='shipped'),
    Row(order_id=3, customer_id=101, order_date=date(2026, 1, 7), amount=240.00, status='pending'),
    Row(order_id=4, customer_id=103, order_date=date(2026, 1, 8), amount=18.99,  status='cancelled'),
    Row(order_id=5, customer_id=104, order_date=date(2026, 1, 9), amount=999.95, status='shipped'),
]
df = spark.createDataFrame(rows)
df.show()
```

```python
df.write.mode('overwrite').format('delta').saveAsTable('orders_bronze')
print('Wrote orders_bronze to the Lakehouse.')
```
